```python
import numpy as np
import matplotlib.pylab as plt
```

```python
dt = 0.01
t = np.arange(0, 1, dt)
x = np.cos(t*2*np.pi)

F0 = dt*len(t)
freq = np.fft.fftfreq(len(t), 1/len(t))/F0
cos_f = np.fft.fft(x)

plt.plot(t, x)
plt.show()

plt.stem(freq, cos_f.real*dt)
plt.axis(xmin=-5, xmax=5)
plt.show()
```

```python
dt = 0.01
t = np.arange(0, 1, dt)
x = np.sin(t*2*np.pi)

F0 = dt*len(t)
freq = np.fft.fftfreq(len(t), 1/len(t))/F0
sin_f = np.fft.fft(x)

plt.plot(t, x)
plt.show()

plt.stem(freq, abs(sin_f)*dt)
plt.axis(xmin=-5, xmax=5)
plt.show()

plt.stem(freq, sin_f.real*dt)
plt.axis(xmin=-5, xmax=5)
plt.show()

plt.stem(freq, sin_f.imag*dt)
plt.axis(xmin=-5, xmax=5)
plt.show()
```

```python
dt = 0.01
t = np.arange(0, 2, dt)
square = np.heaviside(t - 1/2, 1) - np.heaviside(t - 3/2, 1)

F0 = dt*len(t)
freq = np.fft.fftfreq(len(t), 1/len(t))/F0
square_f = np.fft.fft(square)

plt.plot(t, square)
plt.show()

plt.stem(freq, abs(square_f)*dt)
plt.axis(xmin=-5, xmax=5)
plt.show()

plt.stem(freq, square_f.real*dt)
plt.axis(xmin=-5, xmax=5)
plt.show()

plt.stem(freq, square_f.imag*dt)
plt.axis(xmin=-5, xmax=5)
plt.show()
```
