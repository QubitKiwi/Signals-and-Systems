```python
import numpy as np
import matplotlib.pylab as plt
```

```python
t = np.arange(-5, 10, 0.01)
x = np.maximum(0, t)
plt.plot(t, x)
plt.show()
```

```python
t = np.arange(-5, 10, 0.01)
x = np.maximum(t+3, 0)
plt.plot(t, x)
plt.show()
```

```python
t = np.arange(-5, 10, 0.01)
x = np.maximum(t+3, 0) - 2*np.maximum(t+2, 0) + np.maximum(t, 0) + np.maximum(t-3, 0) - + np.maximum(t-4, 0)
plt.plot(t, x)
plt.show()
```
