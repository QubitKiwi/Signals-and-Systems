```python
import numpy as np 
import matplotlib.pylab as plt
```

```python
t = np.arange(-3, 3, 0.1)
x = np.heaviside(t, 1)
plt.plot(t, x)
plt.show()
```

```python
t = np.arange(0, 3, 0.01)
x = np.heaviside(t-2, 4)
plt.plot(t, x)
plt.show()
```

```python
t = np.arange(0, 10, 0.01)
x = np.heaviside(t-2, 1) - np.heaviside(t-3, 1) - 2*np.heaviside(t-5, 1) + np.heaviside(t-6, 1) + np.heaviside(t-7, 1)
plt.plot(t, x)
plt.show()
```
